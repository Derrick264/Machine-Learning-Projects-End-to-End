#### Handle Missing Values
Why: Most ML models can’t handle missing values, and missing df2 can bias results.

Numeric columns: Typically filled with the median (robust to outliers).

Categorical/text columns: Filled with 'Unknown' or the mode.

Year: Use the mode (most frequent year).

Target variable (gross): For regression, best to drop rows with missing gross.

```python
missing_counts = df2.isnull().sum()
missing_percent = (missing_counts / len(df2)) * 100
missing_info = pd.DataFrame({'Missing Count': missing_counts, 'Missing %': missing_percent})
print("Columns with Missing Values:\n", missing_info[missing_info['Missing Count'] > 0].sort_values(by='Missing %', ascending=False))
```

```python
# Numeric columns imputation
numeric_cols = [
    'budget', 'gross', 'duration', 'num_critic_for_reviews', 'num_user_for_reviews',
    'director_facebook_likes', 'actor_1_facebook_likes', 'actor_2_facebook_likes',
    'actor_3_facebook_likes', 'cast_total_facebook_likes', 'facenumber_in_poster', 'aspect_ratio'
]

for col in numeric_cols:
    if df2[col].isnull().sum() > 0:
        df2[col].fillna(df2[col].median())
```

```python
# Categorical columns imputation
cat_cols = [
    'director_name', 'actor_2_name', 'actor_3_name', 'plot_keywords', 
    'language', 'country', 'content_rating', 'color'
]
for col in cat_cols:
    if df2[col].isnull().sum() > 0:
        df2[col] = df2[col].fillna('Unknown')
```

```python
# Year column
df2['title_year'] = pd.to_numeric(df2['title_year'], errors='coerce')
df2['title_year'] = df2['title_year'].fillna(df2['title_year'].mode()[0])
df2['title_year'] = df2['title_year'].astype(int)
```

```python
df2.drop_duplicates(inplace=True)
print(f"df2set shape after removing duplicates: {df2.shape}")
```

#### Standardize/Format Text Columns
Why: Uniform formatting helps with encoding, searching, and analysis.

```python
df2['movie_title'] = df2['movie_title'].str.strip()
df2['genres'] = df2['genres'].str.lower().str.replace('|', ', ')
```

Outlier Handling 
Why: Extreme values in budget or gross can dominate model learning. We can cap outliers, remove them, or log-transform.

```python
import numpy as np

# Cap budget and gross to the 99th percentile
for col in ['budget', 'gross']:
    cap = df2[col].quantile(0.99)
    df2[col] = np.where(df2[col] > cap, cap, df2[col])
```

 Encoding Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder

# List categorical columns you want to label encode
cat_cols = ['color', 'language', 'country', 'content_rating']

label_encoders = {}

for col in cat_cols:
    le = LabelEncoder()
    df2[col] = df2[col].astype(str)  # Ensure all categories are strings
    df2[col] = le.fit_transform(df2[col])
    label_encoders[col] = le  # save encoder if you want to inverse transform later
```

 Handling Multi-Label Columns (e.g., Genres)
Movies can belong to multiple genres, separated by commas. Use MultiLabelBinarizer:

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Split genres into lists
df2['genres_list'] = df2['genres'].str.split(', ')

mlb = MultiLabelBinarizer()
genre_dummies = mlb.fit_transform(df2['genres_list'])
genre_df2 = pd.DataFrame(genre_dummies, columns=mlb.classes_)

df2 = pd.concat([df2.drop('genres', axis=1), genre_df2], axis=1)
```

```python
df2.head(3)
```

 Feature Transformation

Why: To correct skewed distributions and make features more comparable.

A. Log Transform Heavily Skewed Variables

```python
import numpy as np

for col in ['gross', 'budget', 'cast_total_facebook_likes', 'num_voted_users']:
    df2[f'log_{col}'] = np.log1p(df2[col])
```

### Feature Engineering

A. Return on Investment (ROI)

```python
df2['roi'] = df2['gross'] / df2['budget']
df2.head(2)
```

B. “Star Power” Score

```python
df2['star_power'] = (
    df2['actor_1_facebook_likes'] +
    df2['actor_2_facebook_likes'] +
    df2['actor_3_facebook_likes'] +
    df2['director_facebook_likes']
)
```

C. Movie Age

```python
current_year = 2025
df2['movie_age'] = current_year - df2['title_year']
```

```python
df2.head(3)
```

Preparing Target Variable :
If you want to create a success class (Hit/Average/Flop), do this now:

```python
def classify_movie(row):
    if row['budget'] == 0:
        return 'Unknown'
    roi = row['gross'] / row['budget']
    if roi < 1:
        return 'Flop'
    elif roi < 2:
        return 'Average'
    else:
        return 'Hit'

df2['success_category'] = df2.apply(classify_movie, axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

def group_rare_and_encode(df, col, threshold=5):
    actor_counts = df2[col].value_counts()
    
    # Plot histogram of actor appearances
    plt.figure(figsize=(8, 4))
    actor_counts.hist(bins=50)
    plt.title(f'Distribution of Number of Movies per Actor: {col}')
    plt.xlabel('Number of Movies')
    plt.ylabel('Number of Actors')
    plt.show()
    
    # Identify rare actors
    rare_actors = actor_counts[actor_counts < threshold].index
    
    # Group rare actors into 'Other'
    df2[col] = df2[col].apply(lambda x: 'Other' if x in rare_actors else x)
    
    # Label encode
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    
    print(f"After grouping rare actors, unique {col} categories:", df[col].nunique())
    
    return df, le

# Usage for actor_1_name and actor_2_name
df1, le_actor_1 = group_rare_and_encode(df2, 'actor_1_name', threshold=5)
df1, le_actor_2 = group_rare_and_encode(df2, 'actor_2_name', threshold=5)
```
